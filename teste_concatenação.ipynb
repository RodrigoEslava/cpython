```python
import pandas as pd
import numpy as np

data_1st = {'a': [114787, 285613], 'b': [3, 4]}
first_df = pd.DataFrame(data=data_1st)

data_2nd = {'114787': ['pqrst', 'klmno'], '285613': ['fghij', 'abcde'], '3': ['uvwxy', 'z01234'], '4': ['!@#$%', '*&^%$']}
second_df = pd.DataFrame(data=data_2nd)
```

```python
result = []
for index, row in first_df.iterrows():
    first_column_from_2nd_df = second_df[str(row['a'])]
    second_column_from_2nd_df = second_df[str(row['b'])]
    resulting_tuple = (first_column_from_2nd_df, second_column_from_2nd_df)
    result.append(resulting_tuple)

first_df['concat'] = result
first_df
```
